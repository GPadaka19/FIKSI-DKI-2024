```python
import os
import pandas as pd
import gdown
from googleapiclient.discovery import build
from googleapiclient.http import MediaIoBaseDownload
from google.oauth2 import service_account
import io

SCOPES = ['https://www.googleapis.com/auth/drive.readonly']
SERVICE_ACCOUNT_FILE = '/content/drive/MyDrive/FIKSI DKI 2024/fiksi-dki-2024-read-only.json'

credentials = service_account.Credentials.from_service_account_file(
        SERVICE_ACCOUNT_FILE, scopes=SCOPES)

drive_service = build('drive', 'v3', credentials=credentials)

file_path = '/content/drive/MyDrive/FIKSI DKI 2024/Proposal.csv'
df = pd.read_csv(file_path)

base_path = '/content/drive/MyDrive/FIKSI DKI 2024'

def get_file_name(file_id):
    file = drive_service.files().get(fileId=file_id, fields='name').execute()
    return file.get('name')

for index, row in df.iterrows():
    bidang = row['Bidang']
    link = row['Link']

    save_path = os.path.join(base_path, bidang)
    os.makedirs(save_path, exist_ok=True)

    file_id = link.split('id=')[1]

    file_name = get_file_name(file_id)

    output = os.path.join(save_path, file_name)

    gdown.download(f'https://drive.google.com/uc?export=download&id={file_id}', output, quiet=False)
```
